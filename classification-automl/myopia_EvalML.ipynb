```python
import evalml
from evalml import AutoMLSearch
from evalml.utils import infer_feature_types
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../content/reduced_filtered_df.csv')
```

```python
X= df.drop('MYOPIC', axis=1).drop('DADMY', axis=1)
y= df['MYOPIC']
```

```python
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X,y,problem_type='binary')
```

```python
from evalml.data_checks import DefaultDataChecks

data_checks = DefaultDataChecks("binary", "log loss binary")
data_checks.validate(X_train, y_train)
```

```python
evalml.problem_types.ProblemTypes.all_problem_types
```

```python
automl = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='binary', objective='auc', additional_objectives=['precision'])
automl.search()
```
